```python
gapminder_copy = gapminder.copy()
```

```python
gapminder_copy.head()
```

```python
gapminder_copy = gapminder_copy.dropna()
#this drops all of the NAN (nulls)
gapminder_copy.head()
```

```python
gapminder_copy = gapminder_copy.drop_duplicates(keep = "first")
#drop_duplicates removes duplicates 
gapminder_copy.head()
```

```python
gapminder_copy.info()
```

```python
gapminder_copy['year'] = gapminder_copy['year'].astype(int)
gapminder_copy['pop'] = gapminder_copy['pop'].astype(int)
```

```python
gapminder_copy.info()
```

```python
gapminder_copy
```

```python
gapminder_copy = gapminder_copy.reset_index(drop=True)
```

```python
gapminder_copy
```

```python
gapminder_copy["region"] = gapminder_copy["region"].str.lstrip()
#removes white space from the left
gapminder_copy["region"] = gapminder_copy["region"].str.rstrip()
#removes white space from the right
gapminder_copy["region"] = gapminder_copy["region"].str.lower()
#makes the letters all lowercase 
```

```python
gapminder_copy
```

```python
gapminder_copy['country'] = gapminder_copy['region'].str.split('_', 1).str[1]
gapminder_copy['continent'] = gapminder_copy['region'].str.split('_', 1).str[0]
gapminder_copy.head()
```

```python
gapminder_copy = gapminder_copy.drop('region', 1)
gapminder_copy.head()
```

```python
gapminder_copy.columns = gapminder_copy.columns.str.lower()
```

```python
gapminder_copy.head()
```

```python
gapminder_copy = gapminder_copy.rename(columns={'life exp' : 'lifeexp'})
```

```python
PRB["region"] = PRB["region"].str.lstrip()
#removes white space from the left
PRB["region"] = PRB["region"].str.rstrip()
#removes white space from the right
PRB["region"] = PRB["region"].str.lower()
```

```python
PRB
PRB["country"] = PRB["region"].str.split("_", 1).str[1].str.lower()
PRB["continent"] = PRB["region"].str.split("_", 1).str[0].str.lower()
PRB = PRB.drop("region", 1)
PRB.head()
```
